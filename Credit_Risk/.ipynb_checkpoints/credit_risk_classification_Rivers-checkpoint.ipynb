### Step 2: Create the labels set (`y`)  from the “loan_status” column, and then create the features (`X`) DataFrame from the remaining columns.

```python
# Separate the data into labels and features
# Separate the features
features = ['loan_size', 'interest_rate', 'borrower_income', 'debt_to_income',
       'num_of_accounts', 'derogatory_marks', 'total_debt']

# The X variable will be the df_scaled DataFrame and is located 3 blocks down.
# The y variable will be df.loan_status and is located 3 blocks down.
```

```python
# Alexander demonstrated scaling this prior to the train_test_split, and I found it worked better to follow along with his demo
# make a subset
df_sub = df.loc[:, features]

# initialize
scaler = StandardScaler()

# fit
scaler.fit(df_sub)

# predict/transform
scaled_data = scaler.transform(df_sub)
df_scaled = pd.DataFrame(scaled_data, columns=features)

df_scaled.head()
```

```python
df_scaled.describe()
```

### Step 3: Split the data into training and testing datasets by using `train_test_split`.

```python
# Train_test_learn module already imported
# Split the data using train_test_split
# Assign a random_state of 1 to the function

# Step 1: Get the data
X = df_scaled
y = df.loan_status

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1, stratify=y) 
# stratify=True maintains target class percentages

# verify the split with .shape for review of the X and y train and test data.
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Create a Logistic Regression Model with the Original Data

###  Step 1: Fit a logistic regression model by using the training data (`X_train` and `y_train`).

### Step 2: Save the predictions on the testing data labels by using the testing feature data (`X_test`) and the fitted model.

### Step 3: Evaluate the model’s performance by doing the following:

* Generate a confusion matrix.

* Print the classification report.

```python
# LogisticRegression module from SKLearn already imported
# This code block sets up the train and test data so that I could run as many classifiers as desired
# The confusion matrix and classification report are added into this code block as well

# Function for Classification
def doClassification(model, X_train, X_test, y_train, y_test):
    # Fit the model using training data
    model.fit(X_train, y_train)
    
    # Evaluate the model
    train_preds = model.predict(X_train)
    test_preds = model.predict(X_test)
    
    train_proba = model.predict_proba(X_train)[:, 1]
    test_proba = model.predict_proba(X_test)[:, 1]

    # Generate metrics TRAIN
    train_cf = confusion_matrix(y_train, train_preds)
    train_cr = classification_report(y_train, train_preds)
    train_auc = roc_auc_score(y_train, train_proba)
    
    train_results = f"""TRAIN METRICS
    Confusion Matrix: 
    {train_cf}
    
    AUC: {train_auc}

    Classification Report:
    {train_cr}
    """
    
    print(train_results)

    # Generate metrics TEST
    test_cf = confusion_matrix(y_test, test_preds)
    test_cr = classification_report(y_test, test_preds)
    test_auc = roc_auc_score(y_test, test_proba)
    
    test_results = f"""TEST METRICS
    Confusion Matrix: 
    {test_cf}
    
    AUC: {test_auc}

    Classification Report:
    {test_cr}
    """
    
    print(test_results)
    
    # VISUALIZE TEST RESULTS
    # Generate a Confusion Matrix
    plt.figure(figsize=(4, 4))
    sns.heatmap(test_cf, annot=True, fmt='d', cmap='Blues', cbar=False, 
                xticklabels=['Predicted Negative', 'Predicted Positive'], 
                yticklabels=['Actual Negative', 'Actual Positive'])
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
    
    # ROC Curve
    fpr, tpr, _ = roc_curve(y_test.values, test_proba)
        
    # Set figure size
    plt.figure(figsize=(6, 6))
        
    # Plot the ROC curve
    plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {test_auc:.2f})')
        
     # Plot the diagonal reference line
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
        
    # Titles and labels
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
        
    # Add grid and legend
    plt.grid(linestyle="--", color="lightgrey", alpha=0.75)
    plt.legend(loc="lower right")
        
    # Display the plot
    plt.show()
```

```python
# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
lr = LogisticRegression(random_state=1)

# Do Machine Learning
doClassification(lr, X_train, X_test, y_train, y_test)
```

```python
# While the assignment is to utilize the Logistic Regression was the required ML model, it is important to try a few different models to determine the best one
# Model #2: Decision Tree Classifier
dt = DecisionTreeClassifier(random_state=1)

# Do Machine Learning
doClassification(dt, X_train, X_test, y_train, y_test)
```

```python
# Model #3: Random Forest Classifier
rf = RandomForestClassifier(random_state=1)

# Do Machine Learning
doClassification(rf, X_train, X_test, y_train, y_test)
```

```python
# Model #4: Support Vector Classification
svc = SVC(probability=True)

# Do Machine Learning
doClassification(svc, X_train, X_test, y_train, y_test)
```

```python
# Model #5: KNeighbors Classifier
# I trialed n_neighbors at 10, 14-16, and 20, and 10 and 15 had the best auc.
knn = KNeighborsClassifier(n_neighbors=15)

# Do Machine Learning
doClassification(knn, X_train, X_test, y_train, y_test)
```
